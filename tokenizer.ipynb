```python
def get_freq(text):
    hash_map={}
    for first,second in zip(text,text[1:]):
        hash_map[(first,second)]=hash_map.get((first,second),0)+1
    return hash_map

def find_most_frequent(hash_map:dict):
    lst=sorted(((v,k) for k,v in hash_map.items()),reverse=True)
    return lst


def merge(ids,pair,idx):
    newids=[]
    i=0
    while(i<len(ids)):
        if i < len(ids)-1 and pair[0]==ids[i] and pair[1]==ids[i+1]:
            newids.append(idx)
            i+=2
        else:
            newids.append(ids[i])
            i+=1
        
    return newids

#top_pair=max(hash_map,key=hash_map.get)
```

```python
vocab_size=276
new_merges=vocab_size-256
ids=list(tokens)
i=0
merges={}
while(i<new_merges):
    freq=get_freq(ids)
    top_pair=max(freq,key=freq.get)
    new_idx=256+i
    ids=merge(ids,top_pair,new_idx)
    merges[top_pair]=new_idx
    i+=1
print(len(ids))
print(len(tokens))
print(len(tokens)/len(ids))
print(merges)
    
```

```python
#Decoder for tokenizer:
vocab={idx :bytes([idx]) for idx in range(256)}
for (p0,p1), idx in merges.items():
    vocab[idx]=vocab[p0]+vocab[p1]


def decode(ids):
    tokens=b"".join(vocab[idx] for idx in ids)
    text=tokens.decode('utf-8', errors='replace')
    return text

print(decode([256]))
```

```python

def encode(text):
    tokens = list(text.encode("utf-8")) # raw bytes
    while len(tokens)>=2:
        freq = get_freq(tokens)
        pair = min(freq, key=lambda p: merges.get(p,float('inf')))
        if pair not in merges:
            break
        idx = merges[pair]
        tokens = merge(tokens,pair, idx)
    return tokens


    return tokens
tokens=encode("hello Working!")
print(encode("hello Working!"))
```
